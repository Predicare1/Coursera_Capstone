```python
import pandas as pd
import numpy as np
```

```python
arrest = pd.read_csv('Arrest_Data_from_2010_to_Present.csv')
```

```python
arrest.head()
```

```python
#Checking the percentage of missing values in the remaining features
total = arrest.isnull().sum().sort_values(ascending=False)
per_miss_val = arrest.isnull().sum()/arrest.isnull().count()*100
per_miss_val = (round(per_miss_val, 2).sort_values(ascending=False))
per_miss_val

#Creating a percentage table for missing values
percent_missing_data = pd.concat([total, per_miss_val], axis=1, keys=['Total', 'Percentage'])
percent_missing_data
```

```python
# Checking information of the data
arrest.info()
```

```python
# Changing the arrest date to a pandas date time
arrest['Arrest Date'] = pd.to_datetime(arrest['Arrest Date'])
```

```python
arrest.tail()
```

```python
# Creating a new column of year
arrest['year'] = arrest['Arrest Date'].dt.year
```

```python
# Filtering to check how many arrests happened in 2018
arrest[arrest['year']==2018].shape[0]
```

```python
# Checking the counts for each year
arrest['year'].value_counts()
```

```python
arrest[arrest['Charge Group Description'].isin(['Vehicle Theft','Robbery', 'Burglary', 'Receive Stolen Property'])]['Age'].quantile(.95)
```

```python
# Creating a new dataframe for just the year 2018
a = arrest[arrest['year']==2018]
# Checking the size of the new dataframe
a.shape
```

```python
# Checking how many missing values are in the dataframee I created
total = a.isnull().sum().sort_values(ascending=False)
total
```

```python
a.head()
```

```python
a[a['Charge Group Description'].isin(['Vehicle Theft','Robbery', 'Burglary', 'Receive Stolen Property'])]['Age'].quantile(.95)
```

```python
# How many bookings of arrestees were made in the area with the most arrests in 2018?
# Doing a value count on Area Name for the year 2018
a['Area Name'].value_counts()
```

```python
# Dropping missing rows of arrests for the Charge Group
a = a.dropna(axis=0, subset=['Charge Group Description'])
a = a.set_index("Charge Group Description")
```

```python
# Dropping rows with 'Pre-Delinquency' and 'Non-Criminal Detention' arrests
a = a.drop(['Pre-Delinquency', 'Non-Criminal Detention'], axis=0)
a.shape # Checking the shape
```

```python
a.head()
```

```python
mean = a.groupby(['Charge Group Description']).Age.transform('mean')
std = a.groupby(['Charge Group Description']).Age.transform('std')
```

```python
z_score = (a.Age - mean)/std
z_score = pd.DataFrame(z_score)
```

```python
z_score.max()
```
